# Data Cleaning

We would have to rename the columns for clarity sake and format the "Day of the week" column to actual names of the days rather than just numbers representing those days

-  Rename some Column
- Change Days of the week from numerical representation to actual name of days

```python
# Rename columns

df.rename(columns = {
    'Week of the month (first week, second, third, fourth or fifth week': 'Week od the Month',
    'Day of the week (Monday to Friday)': 'Day of the Week',
    'Orders from the traffic controller sector': 'Traffic controller orders',
    'Target (Total orders)':'Total orders'
}, inplace = True)
```

```python
df.head()
```

### Name the Days in a Week

```python
days = {
    1: 'Monday',
    2: 'Tuesday',
    3: 'Wednesday',
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday',
    7: 'Sunday'
}
df['Day of the Week'] = df['Day of the Week'].map(days)
```
